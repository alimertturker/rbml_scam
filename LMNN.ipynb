```python
import numpy as np
import pandas as pd
from metric_learn import LMNN
import matplotlib.pyplot as plt 
from sklearn.manifold import TSNE
```

```python
def plot_tsne(X, y, colormap=plt.cm.Paired):
    plt.figure(figsize=(8, 6))

    # clean the figure
    plt.clf()

    tsne = TSNE()
    X_embedded = tsne.fit_transform(X)
    plt.scatter(X_embedded[:, 0], X_embedded[:, 1], c=y, cmap=colormap)

    plt.xticks(())
    plt.yticks(())

    plt.show()
```

```python
data = pd.read_csv("./dataset/IRIS.csv")
X = data.iloc[:, [0, 1, 2, 3]].values
y = data.iloc[:,[4]]
unique_tags = set(y["species"])
mapper = {k: 0 + i for i, k in enumerate(unique_tags )}
y["species"] = y["species"].map(mapper)
Y = y["species"]
```

```python
plot_tsne(X,Y)
lmnn = LMNN(k=5, learn_rate=1e-6)
lmnn.fit(X, Y)
```

```python
X_lmnn = lmnn.transform(X)

plot_tsne(X_lmnn, Y)
```
